**Hashing** is simply passing some data through a formula that produces a result, called a hash. That hash is usually a string of characters and the hashes generated by a formula are always the same length, regardless of how much data you feed into it.

**Functions** associated :

encode() : Converts the string into bytes to be acceptable by hash function.
digest() : Returns the encoded data in byte format.
hexdigest() : Returns the encoded data in hexadecimal format.

```python
#importing hashlib library which contains different types of algorithms.
import hashlib

```

```python
#getting a input string from user
a=input("enter the string to be hashed:")
#hasing the given string using md5 algorithm.
crypt=hashlib.md5(a.encode())
#printing the hashed data in hexadecimal format
print(crypt.hexdigest())
```

**Salting** works by adding an extra secret value to the input, extending the length of the original password.

```python
#imprting hashlib and binascii library for salting process.
import hashlib,binascii

```

```python
#using the pbkdf2_hmac algortihm with correct paramaters to do salting to the given string"hello"
info=hashlib.pbkdf2_hmac('md5',b'password',b'salt',1000)
binascii.hexlify(info)
```
